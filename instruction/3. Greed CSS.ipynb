```python
# HTML

# Чтобы можно было обращаться к ссылкам или, например, разместить слева от них логотип, создаим еще один контейнер - nav-list

    <div class="navbar">
        <div class="nav-list">  # Поместили в него все ссылки
            <a href="index.html">О нас</a>
            <a href="index.html">Прайс</a>
            <a href="index.html">Поставщики</a>
            <a href="index.html">Контакты</a>
            <a href="index.html">Работа</a>
        </div>
    </div>
```

```python
# СSS

# Зададим классу navbar параметры в стилях

.navbar {
    display: flex; # Располагаем содержимое в одну линию
    background-color: rgb(29, 28, 28); # Задаем фоновый цвет
}

# Зададим параметры контейнеру nav-list

.nav-list {
    display: flex; # Располагаем содержимое в одну линию
}

# Обращаемся к ссылкам, находящимся в контейнере nav-list

.nav-list a {
    display: block; # Задаем параметры display: block, чтобы можно было применить отступы и прочие свойства
    padding: 1em; # Задаем внутренние отступы
    text-decoration: none; # Удаляем нижнее подчеркивание ссылок
    color: aliceblue; # Задаем цвет ссылок
}

# Применим псевдокласс hover. Ссылка при наведении курсора меняет цвет

# Псевдоклассы позволяют стилизировать отображение элементов. Записываются через двоеточие без пробела. 

.nav-list a:hover {
    background-color: rgb(70, 70, 70);
}
```

```python
# HTML

# Сделаем навигационную панель адаптивной. Кнопку создадим через тег button.

<div class="navbar">
        <button class="nav-btn">&#9776;</button> # Код &#9776; - это код готовой кнопки "Бургер"
        <div class="nav-list">
            <a href="index.html">О нас</a>
            <a href="index.html">Прайс</a>
            <a href="index.html">Поставщики</a>
            <a href="index.html">Контакты</a>
            <a href="index.html">Работа</a>
        </div>
</div> 
```

```python
# CSS

# Задаем свойства классу кнопки

.nav-btn {
    font-size: 2em; # Двойная высота
    padding-inline: 10px; # Левый и правый отступы
    padding-bottom: 3px; # Нижний отступ
    border: none; # Убираем линию границы
    background-color: rgb(29, 28, 28); # Цвет фона
    color: aliceblue; # Цвет текста
    cursor: pointer; # При наведении курсора на кнопку он отобразится как указатель
}
```

```python
# CSS

# Припишем сначала:

.nav-btn {
    display: none;  # Т.е при обычном расширении экрана эта кнопка не будет отображаться
}

# Создадим медиа-запросы и поместим в них весь код класса .nav-btn, созданный ранее

# Кнопка появится на экране при уменьшении разрешения дисплея

@media (max-width: 1000px) {
    .nav-btn {
        font-size: 2em;
        padding-inline: 10px; 
        padding-bottom: 3px;
        border: none;
        background-color: rgb(29, 28, 28);
        color: aliceblue;
        cursor: pointer;
        display: block;  # Обязательно добавим этот артибут для отображения кнопки
    }
}
```

```python
# Изменим отображение текста ссылок из контейнера .nav-list

# Он не должен отображаться при уменьшении расширения экрана. Будет видна только кнопка-бургер

@media (max-width: 1000px) {
    .nav-btn {
        font-size: 2em;
        padding-inline: 10px;
        padding-bottom: 3px;
        border: none;
        background-color: rgb(29, 28, 28);
        color: aliceblue;
        cursor: pointer; 
        display: block;
    }

    .nav-list {
        display: none;  # При уменьшении разрешения текст ссылок не будет отображаться.
    }
}
```

```python
# Выпадающее меню(из кнопки-бургера)

# HTML

# Внутри контейнера навигационной панели navbar создадим еще один контейнер - drop и поместим в него контейнер с сылками

    <div class="navbar">
        <div class="drop">  # Новый контейнер drop
            <button class="nav-btn">&#9776;</button>
            <div class="nav-list">
                <a href="index.html">О нас</a>
                <a href="index.html">Прайс</a>
                <a href="index.html">Поставщики</a>
                <a href="index.html">Контакты</a>
                <a href="index.html">Работа</a>
            </div>
        </div>
    </div>
```

```python
# Выпадающее меню(из кнопки-бургера)

# CSS

# В медиа-запросах через новый класс drop обратимся к ссылкам в контейнере nav-list.

@media (max-width: 1000px) {
    .nav-btn {
        font-size: 2em;
        padding-inline: 10px;
        padding-bottom: 3px;
        border: none;
        background-color: rgb(29, 28, 28);
        color: aliceblue;
        display: block;
        cursor: pointer;
    }

    .nav-list {
        display: none;
    }
    
    .nav-btn:hover {
        background-color: rgb(70, 70, 70); # При наведении курсора изменится цвет фона
    }

    .drop:hover .nav-list {  # Псевдокласс :hover является доп.условием(при наведении курсора)
        display: block;  # Эти параметры используем, чтобы ссылки проявились на дисплее в блочном контейнере
        position: absolute; # Меню будет выпадать над кнопкой-бургер. Без этого параметра меню выпадает на всю страницу
    }
}

# При использовании параметра absolute, расположение элемента можно корректировать: top:..px, right:..px и т.д
```

```python
# Расположим кнопку-бургер на дисплее справа

# CSS 

# Добавим в медиа-запросы параметры для основного контейнера меню - navbar

@media (max-width: 1000px) {
    .nav-btn {
        font-size: 2em;
        padding-inline: 10px;
        padding-bottom: 3px;
        border: none;
        background-color: rgb(29, 28, 28);
        color: aliceblue;
        display: block;
        cursor: pointer;
    }

    .nav-list {
        display: none;
    }

    .drop:hover .nav-list {
        display: block;
        position: absolute;
        right: 1px; # Сместили положение выпадающего меню вправо на 1px, чтобы оно поместилось на экране
        background-color: rgb(29, 28, 28);
    }

    .navbar {
        justify-content: space-between; # Элементы располагаются на всей ширине экрана на максимальном удалении
    }
}
```
